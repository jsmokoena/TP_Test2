```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('cleaned_household_service_delivery_data.csv')
```

```python
## 1. Bar Chart: Distribution of Household Types (Formal vs Informal)

plt.figure(figsize=(10, 6))
sns.countplot(data=df, x='Housing_Type', palette='viridis')
plt.title('Distribution of Housing Types')
plt.xlabel('Housing Type')
plt.ylabel('Count')
plt.show()
```

```python
## 3. Histogram: Age Distribution of Household Heads


plt.figure(figsize=(10, 6))
plt.hist(df['Head_Age'], bins=20, color='skyblue', edgecolor='black')
plt.title('Age Distribution of Household Heads')
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.show()
```

```python
## 2. Pie Chart: Access to Water Services

water_access_counts = df['Water_Access'].value_counts()
plt.figure(figsize=(8, 8))
plt.pie(water_access_counts, labels=water_access_counts.index, autopct='%1.1f%%', colors=['#4CAF50', '#FFC107'])
plt.title('Access to Water Services')
plt.show()
```

```python
#THE ADVANCED PLOTS

#1 Pair Plot: Relationships between Variables

plt.figure(figsize=(12, 8))
sns.pairplot(df[['Head_Age', 'Household_Size','Head_Gender', 'Satisfaction_Services']], diag_kind='kde', palette='husl')
plt.suptitle('Pair Plot of Selected Variables', y=1.02)
plt.show()
```

```python
#2. Box Plot: Household Size by Housing Type

plt.figure(figsize=(10, 6))
sns.boxplot(data=df, x='Housing_Type', y='Household_Size', palette='viridis')
plt.title('Box Plot of Household Size by Housing Type')
plt.xlabel('Housing Type')
plt.ylabel('Household Size')
plt.show()
```

```python
# Summary Statistics

summary_stats = df.describe(include='all')
print("Summary Statistics:")
print(summary_stats)
```

```python
# Highlighting significant patterns

## 1. Regions with highest population density

population_density = df.groupby('Region')['Household_Size'].sum().sort_values(ascending=False)
print("\nRegions with the highest population density:")
print(population_density.head())

## 2. Regions most lacking in basic services (water access)

water_access_lacking = df[df['Water_Access'] == 'No']['Region'].value_counts()
print("\nRegions most lacking in water access:")
print(water_access_lacking.head())

## 3. Correlation between 'Head_Age' and 'Household_Size'

correlation_head_age_household_size = df[['Head_Age', 'Household_Size']].corr().iloc[0, 1]
print(f"\nCorrelation between Head Age and Household Size: {correlation_head_age_household_size:.2f}")
```
